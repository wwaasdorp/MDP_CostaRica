```python
import numpy as np

# Convert degrees to radians
alpha = np.radians(10)  # Rotation around the y-axis - swing
beta = np.radians(70)     # Rotation around the x-axis - tilt
gamma = np.radians(88)   # Rotation around the z-axis - azimuth

# Rotation matrix around the y-axis (345 degrees)
R_y = np.array([
    [np.cos(alpha), 0, np.sin(alpha)],
    [0, 1, 0],
    [-np.sin(alpha), 0, np.cos(alpha)]
])

# Rotation matrix around the x-axis (10 degrees)
R_x = np.array([
    [1, 0, 0],
    [0, np.cos(beta), -np.sin(beta)],
    [0, np.sin(beta), np.cos(beta)]
])

# Rotation matrix around the z-axis (190 degrees)
R_z = np.array([
    [np.cos(gamma), -np.sin(gamma), 0],
    [np.sin(gamma), np.cos(gamma), 0],
    [0, 0, 1]
])

# Combined rotation matrix (R_z * R_x * R_y)
R_combined = R_z @ R_x @ R_y

# Print the resulting rotation matrix
print("Combined Rotation Matrix (R_z * R_x * R_y):\n", R_combined)
```

```python
import numpy as np

def extract_angles_from_rotation_matrix(R):
    # Azimuth (rotation around z-axis)
    azimuth = np.arctan2(R[1, 0], R[0, 0])
    
    # Swing (rotation around y-axis)
    swing = np.arcsin(-R[2, 0])
    
    # Tilt (rotation around x-axis)
    tilt = np.arctan2(R[2, 1], R[2, 2])
    
    # Convert from radians to degrees
    azimuth_deg = np.degrees(azimuth)
    swing_deg = np.degrees(swing)
    tilt_deg = np.degrees(tilt)
    
    return azimuth_deg, swing_deg, tilt_deg

# Example rotation matrix (R_combined from previous step)

# Extract the angles
azimuth, swing, tilt = extract_angles_from_rotation_matrix(R_combined)

def new_azimuth(az):
    az = -az
    if az < 0:
        az += 360
    return az

# Print the results
print(f"Azimuth: {new_azimuth(azimuth):.2f} degrees")
print(f"Swing: {swing:.2f} degrees")
print(f"Tilt: {tilt:.2f} degrees")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given angles for plotting
azimuth_deg = azimuth  # degrees
swing_deg = swing      # degrees
tilt_deg = tilt        # degrees

# Convert angles from degrees to radians
azimuth_rad = np.radians(azimuth_deg)
swing_rad = np.radians(swing_deg)
tilt_rad = np.radians(tilt_deg)

# Calculate the rotation matrix for each angle
# Rotation around z-axis (azimuth)
R_z = np.array([
    [np.cos(azimuth_rad), -np.sin(azimuth_rad), 0],
    [np.sin(azimuth_rad), np.cos(azimuth_rad), 0],
    [0, 0, 1]
])

# Rotation around y-axis (swing)
R_y = np.array([
    [np.cos(swing_rad), 0, np.sin(swing_rad)],
    [0, 1, 0],
    [-np.sin(swing_rad), 0, np.cos(swing_rad)]
])

# Rotation around x-axis (tilt)
R_x = np.array([
    [1, 0, 0],
    [0, np.cos(tilt_rad), -np.sin(tilt_rad)],
    [0, np.sin(tilt_rad), np.cos(tilt_rad)]
])

# Combined rotation matrix (R_z * R_y * R_x)
R_combined = R_z @ R_y @ R_x

# Create a grid of points (for plotting the original axes)
length = 2
origin = np.array([0, 0, 0])
x_axis = np.array([[0, length], [0, 0], [0, 0]])
y_axis = np.array([[0, 0], [0, length], [0, 0]])
z_axis = np.array([[0, 0], [0, 0], [0, length]])

# Apply the combined rotation to the axes
x_axis_rotated = R_combined @ x_axis
y_axis_rotated = R_combined @ y_axis
z_axis_rotated = R_combined @ z_axis

# Plotting
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot original axes in black
ax.plot(x_axis[0], x_axis[1], x_axis[2], color='black', linewidth=2, label='X-axis')
ax.plot(y_axis[0], y_axis[1], y_axis[2], color='black', linewidth=2, label='Y-axis')
ax.plot(z_axis[0], z_axis[1], z_axis[2], color='black', linewidth=2, label='Z-axis')

# Plot rotated axes in red, green, and blue
ax.plot(x_axis_rotated[0], x_axis_rotated[1], x_axis_rotated[2], color='red', linewidth=2, label='Rotated X-axis')
ax.plot(y_axis_rotated[0], y_axis_rotated[1], y_axis_rotated[2], color='green', linewidth=2, label='Rotated Y-axis')
ax.plot(z_axis_rotated[0], z_axis_rotated[1], z_axis_rotated[2], color='blue', linewidth=2, label='Rotated Z-axis')

# Set limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])

# Labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Rotation Visualization: Azimuth, Swing, and Tilt')
ax.legend()
ax.grid()

# Show plot
plt.show()
```
